### 04. Missing values

```python
#Check for missing values

print(df.isnull().sum())
```

```python
#Visualize missing values

plt.figure(figsize=(12, 8))
sns.heatmap(df.isnull(), cbar=False)
```

```python
# Identify missing values in the dataset
missing_values = df.isnull().sum()

# Calculate the percentage of missing values for each column
missing_percentage = (missing_values / len(df)) * 100

# Display the percentage of missing values for each column
missing_percentage.sort_values(ascending=False)
```

```python
# Removing columns with 100% missing values and internal codes not relevant to this analysis
columns_to_remove = ['Geo point', 'ID_DIVISION', 'ID_ATELIER_HORTICOLE', 'IDA3D_ENB', 'SITE_VILLES', 'ID_EQPT', 'Compétence']
df_cleaned = df.drop(columns=columns_to_remove)

# Check the updated dataset
df_cleaned.head()
```

```python
# Checking the current data types of each column
df_cleaned.dtypes
```

```python
# Converting data types where necessary
# Example: Converting 'Identifiant espace vert' to integer (if no NaN values) and 'Code postal' to string
df_cleaned['Identifiant espace vert'] = pd.to_numeric(df_cleaned['Identifiant espace vert'], errors='coerce', downcast='integer')
df_cleaned['Code postal'] = df_cleaned['Code postal'].astype(str).str.replace('.0', '', regex=False)

# Display the updated data types
df_cleaned.dtypes
```

```python
# Trimming leading and trailing spaces in string columns
string_columns = df_cleaned.select_dtypes(include='object').columns

for col in string_columns:
    df_cleaned[col] = df_cleaned[col].str.strip()

# Display a few rows to verify the changes
df_cleaned.head()
```

```python
# Examining categorical data for inconsistencies
categorical_columns = df_cleaned.select_dtypes(include='object').columns

# Display unique values for each categorical column to check for inconsistencies
categorical_inconsistencies = {col: df_cleaned[col].unique() for col in categorical_columns}
categorical_inconsistencies
```

```python
# Examining numerical columns for obvious outliers
numerical_columns = df_cleaned.select_dtypes(include=['float64', 'int64']).columns

# Display summary statistics for numerical columns
numerical_summary = df_cleaned[numerical_columns].describe()
numerical_summary
```

```python
# Count the number of entries where 'Année de l'ouverture' is 2025 and above

years_above_2024_count = df_cleaned[df_cleaned["Année de l'ouverture"] >= 2025].shape[0]
years_above_2024_count
```

```python
# Count the number of entries where 'Année de l'ouverture' is exactly 9999

year_9999_count = df_cleaned[df_cleaned["Année de l'ouverture"] == 9999].shape[0]
year_9999_count
```

```python
# Scatterplot to visualize outliers in the 'Année de l'ouverture' column
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned, x='Identifiant espace vert', y='Année de l\'ouverture', alpha=0.6)

plt.title('Scatterplot of Année de l\'ouverture')
plt.xlabel('Identifiant Espace Vert')
plt.ylabel('Année de l\'ouverture')
plt.show()
```

```python
# Omitting the years listed as 9999 from the scatterplot
filtered_data = df_cleaned[df_cleaned["Année de l'ouverture"] != 9999]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=filtered_data, x='Identifiant espace vert', y="Année de l'ouverture", alpha=0.6)

plt.title("Scatterplot of Année de l'ouverture (Excluding 9999)")
plt.xlabel('Identifiant Espace Vert')
plt.ylabel("Année de l'ouverture")
plt.show()
```

```python
# Count the number of unique 'Code postal' in the dataset
unique_postal_codes_count = df_cleaned['Code postal'].nunique()
unique_postal_codes_count
```

```python
# Creating a scatterplot with 'Code postal' instead of 'Identifiant espace vert'
plt.figure(figsize=(12, 8))
sns.scatterplot(data=filtered_data, x='Code postal', y="Année de l'ouverture", alpha=0.6)

plt.title("Scatterplot of Année de l'ouverture by Code Postal (Excluding 9999)")
plt.xlabel('Code Postal')
plt.ylabel("Année de l'ouverture")
plt.xticks(rotation=45)
plt.show()
```

```python
# Filtering the dataset to show the years between 2025 and 9999
years_between_2025_and_9999 = df_cleaned[(df_cleaned["Année de l'ouverture"] >= 2025) & 
                                           (df_cleaned["Année de l'ouverture"] < 9999)]

years_between_2025_and_9999["Année de l'ouverture"].unique()
```

#### The only year between 2025 and 9999 (exclusive of 9999) in the 'Année de l'ouverture' (year of opening) column is 9998. This suggests that apart from the year 9999, which is likely a placeholder or erroneous value, the year 9998 is also present and may require similar consideration.

### 06. Formatting for consistency

```python
# Creating a bar plot for the frequency of 'Catégorie' arranged in alphabetical order

cat_freq_alpha = df_cleaned['Catégorie'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.barplot(x=cat_freq_alpha.values, y=cat_freq_alpha.index)

plt.title("Bar Plot of Frequency of Catégorie (Alphabetical Order)")
plt.xlabel('Frequency')
plt.ylabel('Catégorie')
plt.show()
```

```python
# Formatting spellings for consistency in the 'Catégorie' column

consistent_categories = {
    'Jardinere': 'Jardiniere', 'Jardinieres': 'Jardiniere', 'Jardinières': 'Jardiniere',
    'jardineres': 'Jardiniere', 'jardiniere': 'Jardiniere', 'jardinieres': 'Jardiniere',
    'jardinièes': 'Jardiniere', 'jardinières': 'Jardiniere', 'jardinères': 'Jardiniere',
    'square': 'Square'
}

df_cleaned['Catégorie'] = df_cleaned['Catégorie'].replace(consistent_categories)

# Checking the updated values
df_cleaned['Catégorie'].unique()
```

```python
# Creating a bar plot for the frequency of 'Catégorie' with formatted names and arranged in ascending order

formatted_cat_freq = df_cleaned['Catégorie'].value_counts().sort_values()
plt.figure(figsize=(12, 8))
sns.barplot(x=formatted_cat_freq.values, y=formatted_cat_freq.index)

plt.title("Bar Plot of Frequency of Catégorie (Formatted and Ascending Order)")
plt.xlabel('Frequency')
plt.ylabel('Catégorie')
plt.show()
```
